##2. Data Loading and Preprocessing
loaded the file paths into a Pandas DataFrame for easy analysis and visualization.

```python
import pandas as pd
import glob
import os

#function to load all the datsets ^*||*^
def load_data_to_def(data_dir):
  real_images = glob.glob(os.path.join(data_dir, 'REAL', '*.*'))
  fake_images = glob.glob(os.path.join(data_dir, 'FAKE', '*.*'))
#labeling class id for real and fake as 0 and 1
  real_df = pd.DataFrame({'filepath':real_images,'label':'Real', 'class_id': 0})
  fake_df = pd.DataFrame({'filepath': fake_images,'label':'AI Generated', 'class_id': 1})
  return pd.concat([real_df, fake_df],ignore_index=True)

train_df = load_data_to_def(train_dir)
test_df = load_data_to_def(test_dir)

#just showing the total image avail in the direc.

print(f"Total training images:{len(train_df)}")
print(f"Total testing images: {len(test_df)}")
train_df.head()
```

#3. environment setup for EDA(Exploratory Data Analysis)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import numpy as np

plt.style.use('dark_background')
sns.set_palette("husl")
```

##VISUALISATION

##*I*. class distribution (img counts are balanced or not)
basically it represents the count of images for real and fake and for training equal image are required.

```python

train_df['label'].value_counts().plot(kind='bar')
plt.title("Number of Images per Class")

plt.xlabel("Class")
plt.ylabel("Count")


plt.show()
```

##*II.* sample real image grid (random img)
showing grid of random real images sapmle

```python
def plot_sample_grid(df, label, n=9):
    paths = df[df['label'] == label]['filepath'].sample(n)

    plt.figure(figsize=(7,7))

    for i, path in enumerate(paths):
#opncv loads in bgr,matplotlib needs rgb so convert into rgb *^||^*

        img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)

        plt.subplot(3, 3, i+1)
        plt.imshow(img) #image show
        plt.axis('off')

    plt.suptitle(f"sample {label} Images")
    plt.tight_layout()
    plt.show() #grid show

plot_sample_grid(train_df, 'Real')
```

##*III.* sample AI image grid(random img)

```python
plot_sample_grid(train_df,'AI Generated')
```

##*IV.* avg brightness for ai and real img classification

```python

train_df['brightness'] = train_df['filepath'].apply(
    lambda x: np.mean(cv2.imread(x, cv2.IMREAD_GRAYSCALE))
)
sns.histplot(data=train_df, x='brightness', hue='label', bins=30)

plt.title("Brightness Distribution")
plt.show()
```

##*V.* image widths distribution (x=px,y=count)

```python
sns.histplot(
    train_df['filepath'].apply(lambda x: cv2.imread(x).shape[1]),
    bins=30
)
plt.show()
```
