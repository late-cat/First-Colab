## ML and data analysis [AI v/s real photos detection]

this note book walks through data downloading, preprocessing, exploratory data analysis with 10 meaningful visualizations, and developing a Convolutional Neural Network (CNN) to differentiate between Real and AI-generated photographs.

## 1. setup and dataset download from kaggle.
used CFAKE dataset from kaggle

```python
!pip install kagglehub
import kagglehub
import os
dataset_path=kagglehub.dataset_download("birdy654/cifake-real-and-ai-generated-synthetic-images")
train_dir=os.path.join(dataset_path,'train')
test_dir=os.path.join(dataset_path,'test')
print("Train directory:",train_dir)
print("Test directory:", test_dir)
```

##2. Data Loading and Preprocessing
loaded the file paths into a Pandas DataFrame for easy analysis and visualization.

```python
import pandas as pd
import glob
import os

#function to load all the datsets ^*||*^
def load_data_to_def(data_dir):
  real_images = glob.glob(os.path.join(data_dir, 'REAL', '*.*'))
  fake_images = glob.glob(os.path.join(data_dir, 'FAKE', '*.*'))
#labeling class id for real and fake as 0 and 1
  real_df = pd.DataFrame({'filepath':real_images,'label':'Real', 'class_id': 0})
  fake_df = pd.DataFrame({'filepath': fake_images,'label':'AI Generated', 'class_id': 1})
  return pd.concat([real_df, fake_df],ignore_index=True)

train_df = load_data_to_def(train_dir)
test_df = load_data_to_def(test_dir)

#just showing the total image avail in the direc.

print(f"Total training images:{len(train_df)}")
print(f"Total testing images: {len(test_df)}")
train_df.head()
```

##*II.* sample real image grid (random img)
showing grid of random real images sapmle

```python
def plot_sample_grid(df, label, n=9):
    paths = df[df['label'] == label]['filepath'].sample(n)

    plt.figure(figsize=(7,7))

    for i, path in enumerate(paths):
#opncv loads in bgr,matplotlib needs rgb so convert into rgb *^||^*

        img = cv2.cvtColor(cv2.imread(path), cv2.COLOR_BGR2RGB)

        plt.subplot(3, 3, i+1)
        plt.imshow(img) #image show
        plt.axis('off')

    plt.suptitle(f"sample {label} Images")
    plt.tight_layout()
    plt.show() #grid show

plot_sample_grid(train_df, 'Real')
```

##*III.* sample AI image grid(random img)

```python
plot_sample_grid(train_df,'AI Generated')
```

##*VI.* brighness distribution using boxplot

##*VII.* brightness range for each class in bar

```python
sample_df = train_df.sample(1000)

sample_df['brightness'] = sample_df['filepath'].apply(
    lambda x: np.mean(cv2.imread(x, 0))
)

sample_df.groupby('label')['brightness'].agg(['min', 'max']).plot(kind='bar')
plt.title("Brightness Range per Class")

plt.show()
```

##*VIII.* edge density for each class
generally natural imgaes has more edge density(texture,noise) than AI images

```python
sample_df = train_df.sample(1000)

sample_df['edge_density'] = sample_df['filepath'].apply(
    lambda x: np.mean(cv2.Canny(cv2.imread(x, 0), 100, 200))
)

sample_df.groupby('label')['edge_density'].mean().plot(kind='bar')
plt.show()
```

##*IX.* Average RGB Channel Values per Class
compares the average red, green, and blue channel intensities between real and AI-generated images.

```python
sample_df = train_df.sample(1000)

sample_df[['avg_red','avg_green','avg_blue']] = sample_df['filepath'].apply(
    lambda x: pd.Series(
        np.mean(img[:,:,::-1], axis=(0,1)) if (img := cv2.imread(x)) is not None else [None]*3
    )
)

sns.heatmap(
    sample_df.groupby('label')[['avg_red','avg_green','avg_blue']].mean(),
    annot=True
)

plt.show()
```

## ***X.*** Image Sharpness (boxplot)
generally ai genrated img are too smooth

```python
sample_df = train_df.sample(1000)

sample_df['blur'] = sample_df['filepath'].apply(
    lambda x: cv2.Laplacian(cv2.imread(x, 0), cv2.CV_64F).var()
)

sns.boxplot(x='label', y='blur', data=sample_df)
plt.show()
```

```python
import tensorflow as tf
train_ds = tf.keras.utils.image_dataset_from_directory(train_dir, image_size=(32,32), batch_size=256)
test_ds = tf.keras.utils.image_dataset_from_directory(test_dir, image_size=(32,32), batch_size=256)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


model = keras.Sequential([
    layers.Rescaling(1./255, input_shape=(32, 32, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(2)
])


model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

#Train
history = model.fit(
    train_ds,
    epochs=5,
    validation_data=test_ds
)
```

### Evaluate Model Performance
Let's evaluate the trained model on the `test_ds` to get a clearer picture of its accuracy and loss on unseen data. This will help us understand if the bias you're observing is systemic.

```python
test_loss, test_accuracy = model.evaluate(test_ds)
print(f"\nTest Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['acc','val_acc'])
plt.show()
```
